```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file1_path = os.path.join('data', 'clinicaltrial_data.csv')
file2_path = os.path.join('data', 'mouse_drug_data.csv')

mouse_drug_df = pd.read_csv(file1_path)
clinical_trial_df = pd.read_csv(file2_path)

merged_df = pd.merge(mouse_drug_df, clinical_trial_df, on='Mouse ID')
df = merged_df.fillna(0)
df = merged_df
df.head()
```

## Tumor Response to Treatment

```python
drug_df = pd.pivot_table(df, 
                        index='Timepoint',
                        columns='Drug', 
                        values='Tumor Volume (mm3)',                        
                        aggfunc=np.mean)

drug_df.head()
```

```python
colors = ['red', 'green', 'orange', 'blue', 'yellow', 'brown', 'purple', 'pink', 'black', 'grey']
markers = ['^', 'o', 's', '*', 'D', 'x', 'P', '2', '$...$', 'H']

x = drug_df.index 
for column, i in zip(drug_df.columns, range(len(drug_df.columns))):
    y = drug_df[column]
    plt.errorbar(x, y, yerr=3, elinewidth=0.8, 
                 color=colors[i], marker=markers[i], 
                 linestyle=':', linewidth=1,
                 errorevery=1, label=column)
    
plt.xlabel('Time(Days)')
plt.ylabel('Tumor Volume (mm3)')
plt.title('Tumor Response to Treatment')
plt.legend()
plt.xlim(-5, max(x) + 5)
plt.ylim(32, max(y) + 5)
plt.grid(axis='y', linestyle='-')
plt.savefig('Fig1 Treatment Plot.png')
plt.show()
```

## Metastatic Response to Treatment

```python
df.head()
```

```python
meta_df = pd.pivot_table(df, 
                        index='Timepoint',
                        columns='Drug', 
                        values='Metastatic Sites',                        
                        aggfunc=np.mean)

# For a clearer view, only 4 Drugs will be sampled for analysis
meta_df = meta_df.drop(['Naftisol', 'Propriva', 'Ramicane', 'Stelasyn', 'Zoniferol', 'Ceftamin'],axis=1)

meta_df.head()
```

```python
colors = ['red', 'green', 'orange', 'blue']
markers = ['^', 'o', 's', '*']

x = meta_df.index
    
for column, i in zip(meta_df.columns, range(len(meta_df.columns))):
    y = meta_df[column]
    plt.errorbar(x, y, yerr=0.2, elinewidth=0.8, 
                 color=colors[i], marker=markers[i], 
                 linestyle=':', linewidth=1,
                 errorevery=1, label=column)
    
plt.xlabel('Treatment Duration(Days)')
plt.ylabel('Met. Sites')
plt.title('Metastatic Spread During Treatment')
plt.legend()
plt.xlim(-5, max(x) + 5)
plt.ylim(-0.2, max(y) + 0.5)
plt.grid(b=True, axis='y')
plt.savefig('Fig2 Meta Spread.png')
plt.show()
```

## Survival Rates

```python
df.head()
```

```python
survival_df = pd.pivot_table(df, 
                            index='Timepoint',
                            columns='Drug', 
                            values='Mouse ID',
                            aggfunc='count')

survival_df = survival_df.drop(['Naftisol', 'Propriva', 'Ramicane', 'Stelasyn', 'Zoniferol', 'Ceftamin'],axis=1)
survival_df
```

```python
colors = ['red', 'green', 'orange', 'blue']
markers = ['^', 'o', 's', '*']

x = survival_df.index
 
for column, i in zip(survival_df.columns, range(len(survival_df.columns))):
    y = (survival_df[column] / survival_df[column][0]) * 100
    plt.errorbar(x, y, yerr=0.2, elinewidth=0.8, 
                 color=colors[i], marker=markers[i], 
                 linestyle=':', linewidth=1,
                 errorevery=1, label=column)
  
plt.xlabel('Treatment Duration(Days)')
plt.ylabel('Survival Rate (%)')
plt.title('Survival Rate During Treatment')
plt.legend()
plt.xlim(-2, max(x) + 5)
plt.ylim(15, max(y) + 5)
plt.grid(b=True, axis='both')
plt.savefig('Fig3 Survival Rate.png')
plt.show()
```
